# Model Performance

```python
# Imports
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import pickle

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.cluster import KMeans


from sklearn.tree import DecisionTreeClassifier
from sklearn.multioutput import MultiOutputClassifier

from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

import matplotlib.pyplot as plt
```

### Running KMeans to Label Data 

```python
# Load in the embeddings 
path = "../data/X20_embeddings.csv.zip"
df = pd.read_csv(path)
```

```python
# Train the Kmeans clustering model with 8 clusters (from Module 2)
kmeans = KMeans(n_clusters = 8, random_state = 1)
kmeans = kmeans.fit(df)

# Grab the labels from GPT API from Module 2
labels = {
    1: ['social issues', 'personal development', 'business and economics', 'community building'],
    2: ['india', 'updates', 'testing', 'fatalities', 'recoveries', 'healthcare'],
    3: ['face masks', 'safety', 'protection', 'public health', 'prevention'],
    4: ['social media', 'resilience', 'community support', 'online events'],
    5: ['global', 'cases', 'deaths', 'statistics'],
    6: ['politics', 'government response', 'public health', 'conspiracy', 'human rights'],
    7: ['health', 'information','vacccine', 'public awareness'],
    8: ['layoffs', 'misinofrmation', 'mental health', 'lockdown', 'access', 'financial impact', 'political response', 'education']
}
```

```python
# Add the labels as ground truth
df['cluster'] = kmeans.predict(df)

# Add a column caled labels which is a list of strings
df['labels'] = df['cluster'].apply(lambda x: labels[x+1])
```

### Sample 5k Rows

```python
# Sample Rows
df_sampled = df.sample(n = 5_000, random_state = 1)
```

```python
# Split the data 70/30
train, test = train_test_split(df_sampled, test_size = 0.3)

# Remove non-training cols and split to X,y
drop_cols = ['cluster', 'labels']
X_train, X_test = train.drop(drop_cols, axis = 1), test.drop(drop_cols, axis = 1)

# One-hot encode the list of labels to multioutputs
y_train = train['labels'].str.get_dummies(sep =',')
y_test = test['labels'].str.get_dummies(sep =',')
```

### Logistic Regression

```python
# Logistic Regression
lr = MultiOutputClassifier(LogisticRegression(), n_jobs = -1)
lr.fit(X_train, y_train)
pickle.dump(lr, open('../trained_models/lr.pkl', 'wb'))
```

```python
# LR Performance
lr_train_acc = accuracy_score(lr.predict(X_train), y_train)
lr_test_acc = accuracy_score(lr.predict(X_test), y_test)
print(f'LR Train Acc: {lr_train_acc:.4f}')
print(f'LR Test Acc: {lr_test_acc:.4f}')
```

### Linear Discriminent Analysis

```python
# LDA 
lda = MultiOutputClassifier(LinearDiscriminantAnalysis(), n_jobs = -1)
lda.fit(X_train, y_train)
pickle.dump(lda, open('../trained_models/lda.pkl', 'wb'))
```

```python
# LDA Performance
lda_train_acc = accuracy_score(lda.predict(X_train), y_train)
lda_test_acc = accuracy_score(lda.predict(X_test), y_test)
print(f'LDA Train Acc: {lda_train_acc:.4f}')
print(f'LDA Test Acc: {lda_test_acc:.4f}')
```

### Decision Tree Classifier

```python
# DT with 5-Fold CV 
grid = {'estimator__min_samples_split': [2, 4, 6, 8, 10, 12, 14]}

# Create the grid search cross-validation object
dt = MultiOutputClassifier(DecisionTreeClassifier(), n_jobs = -1)
dt_grid_search = GridSearchCV(estimator = dt, param_grid = grid,  cv = 3)
dt_grid_search.fit(X_train, y_train)
```

```python
# Plot CV Results 
fig, ax = plt.subplots(figsize = (8,4))
plt.plot(dt_grid_search.cv_results_['mean_test_score'])
plt.xlabel('Pruning Parameter (CCP Alpha')
plt.ylabel('Mean CV Score')
plt.title('[ Decision Tree ] Mean Cross-Validation Score vs CCP Alpha');
```

```python
# Save best estimator
dt_best_estimator = dt_grid_search.best_estimator_
pickle.dump(dt_best_estimator, open('../trained_models/dt_best_estimator.pkl', 'wb'))
```

```python
# DT Pperformance 
dt_train_acc = accuracy_score(dt_best_estimator.predict(X_train), y_train)
dt_test_acc = accuracy_score(dt_best_estimator.predict(X_test), y_test)
print(f'Decision Tree Train Acc: {dt_train_acc:.4f}')
print(f'Decision Tree Test Acc: {dt_test_acc:.4f}')
```

### Random Forest Classifier 

```python
# RF with 3_ Fold CV 
grid = {'estimator__max_depth': [2, 4, 6, 8]}

# Create the grid search cross-validation object
rf = MultiOutputClassifier(RandomForestClassifier(verbose = 1), n_jobs = -1)
rf_grid_search = GridSearchCV(estimator = rf, param_grid = grid,  cv = 3)
rf_grid_search.fit(X_train, y_train)
```

```python
# RF Performance
rf_train_acc = accuracy_score(rf.predict(X_train), y_train)
rf_test_acc = accuracy_score(rf.predict(X_test), y_test)
print(f'Random Forest Train Acc: {rf_train_acc:.4f}')
print(f'Random Forest Test Acc: {rf_test_acc:.4f}')
```
